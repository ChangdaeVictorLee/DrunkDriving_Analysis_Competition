### 2. 카카오 api를 활용해 x,y좌표로 주소 변환하기

```python
import json
import requests
```

```python
# 카카오 api 키
api_key = 'caeb1715d26238a57b3604621bf1953e'
```

```python
# 좌표 -> 주소
def lat_lon_to_addr(lon,lat):
    url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={longitude}&y={latitude}'.format(longitude=lon,latitude=lat)
    headers = {"Authorization": "KakaoAK " + api_key}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address_name']
    return str(match_first)
```

```python
lat_lon_to_addr(127.394815, 36.344215)
```

### 3. 변환한 주소를 df_cleaned 행정구역에 삽입하기

```python
import time
```

```python
# 한 번 터져서 다시 실행했습니다.
cnt = 0
for i in range(len(df_cleaned)):
    if df_cleaned['행정구역'][i] == 0:
        x = df_cleaned['x'][i]
        y = df_cleaned['y'][i]
        
        try:
            location = lat_lon_to_addr(x,y)
        
        except:
            time.sleep(10)
            location = lat_lon_to_addr(x,y)            
        
        # 주소 삽입하기
        df_cleaned['행정구역'][i] = location
        cnt +=1
print(f'총 {cnt}건의 좌표를 주소로 변환했습니다')
```
